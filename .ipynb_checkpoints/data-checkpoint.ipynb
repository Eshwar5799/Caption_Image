```python
def load_text(path):
    with open(path) as f:
        captions=f.read()
    return captions
```

```python
captions=load_text("G:\Image-Captioning Dataset\Flickr8k.lemma.token.txt")
captions=captions.split('\n')[:-1]
```

```python
# Format Image name #i<caption> where 0<i<=4
```

```python
# Now, we create a dictionary named “descriptions” 
#which contains the name of the image (without the .jpg extension) 
#as keys and a list of the 5 captions for the corresponding image as values.
```

```python
descriptions={}
for line in captions:
    token=line.split()
    image_id,image_desc=token[0],token[1:]
    image_id=image_id.split('.')[0]
    image_desc=' '.join(image_desc)
    if descriptions.get(image_id) is None:
        descriptions[image_id]=list()  
    descriptions[image_id].append(image_desc)
```

```python
# Data cleaning
```

```python

import re
def clean_text(sentence):
    sentence=sentence.lower()
    sentence=re.sub("[^a-z]+"," ",sentence)
    sentence=sentence.split()
    sentence=[s for s in sentence if len(s)>1]
    sentence=" ".join(sentence)
    
    return sentence
```

```python
for key,caption_list in descriptions.items():
    for i in range(len(caption_list)):
        caption_list[i]=clean_text(caption_list[i])
```

```python
with open("G:/Image-Captioning Dataset/cleaned_descriptions.txt","w") as f:
    f.write(str(descriptions))
```

```python
# Using google's Imagenet Architecture
def preprocess(img):
    img=image.load_img(img,target_size=(224,224))
    image_array=image.img_to_array(img)
    image_array=np.expand_dims(image_array,axis=0)
    image_array=preprocess_input(image_array)
    preds=model.predict(image_array)
    return preds

def img_show(img):
    img=cv2.imread(img)
    plt.axis("off")
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    
    return img

def normal_preprocess(img):
    
    img=image.load_img(img,target_size=(224,224))
    image_array=image.img_to_array(img)
    image_array=np.expand_dims(image_array,axis=0)
    image_array=preprocess_input(image_array)
    return image_array 
    
```

```python
word_to_idx={}
idx_to_word={}
for i,word in enumerate(total_words):
    word_to_idx[word]=i+1
    idx_to_word[i+1]=word
```

```python
vocab_size=len(word_to_idx)+1
print("Vocab size is:%d"%(vocab_size))
```

```python
# Calculating the maximum length of caption
def all_lines(descriptions):
    all_desc=[]
    for key in descriptions.keys():
        [all_desc.append(d) for d in descriptions[key]]
    return all_desc

def max_length(descriptions):
    lines=all_lines(descriptions)
    value= max(len(d.split()) for d in lines)
    return value
```

```python
maximum_length=max_length(train_dictionary)
print("The maximum length of caption is: %d" %(maximum_length))
```

For batch processing
we need to make sure that each sequence is of equal length.
Hence we need to append 0’s (zero padding) at the end of each sequence.




Data Generator so that we can feed data to the deep learning model.
Load the Images and Captions in Batches

```python
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
def data_generator(descriptions, photos, wordtoix, max_length, num_photos_per_batch):
    X1, X2, y = list(), list(), list()
    n=0
    # loop for ever over images
    while 1:
        for key, desc_list in descriptions.items():
            n+=1
            # retrieve the photo feature
            photo = photos[key]
            for desc in desc_list:
                # encode the sequence
                seq = [wordtoix[word] for word in desc.split(' ') if word in wordtoix]
                # split one sequence into multiple X, y pairs
                for i in range(1, len(seq)):
                    # split into input and output pair
                    in_seq, out_seq = seq[:i], seq[i]
                    # pad input sequence
                    in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
                    # encode output sequence
                    out_seq = to_categorical([out_seq], num_classes=vocab_size)[0]
                    # store
                    X1.append(photo)
                    X2.append(in_seq)
                    y.append(out_seq)
            # yield the batch data
            if n==num_photos_per_batch:
                yield [[array(X1), array(X2)], array(y)]
                X1, X2, y = list(), list(), list()
                n=0
```

                                    Word Embeddings
-Map every word(index) to a 200-long vector


-Pre Trained glove Model

```python
f=open("C:/Users/Eshwar/Anaconda_latest/Lib/site-packages/gensim/test/test_data/test_glove.txt",
          "r",encoding='utf8') 
embedding_index={}

for line in f:
    values=line.split()
    captions=values[0]
    coefs=np.array(values[1:],dtype='float')
    embedding_index[captions]=coefs
f.close()
    
```

                                   Embedding Matrix

```python
embedding_dim=50
def embedding_matrix():
    
    
    matrix=np.zeros((vocab_size,embedding_dim))
    for word,idx in word_to_idx.items():
        embedding_vector=embedding_index.get(word)
        if embedding_vector is not None:
            matrix[idx]=embedding_vector
    return matrix
```

```python
get_embedding_matrix=embedding_matrix()
print(get_embedding_matrix)
```

                                Model Architecture
                                
  

```python
from keras import Input
from keras.layers import Dropout,Dense,Embedding,LSTM,add
import keras.optimizers 
# Input 1
inputs1=Input(shape=(2048,))
fe1=Dropout(0.5)(inputs1)
fe2=Dense(256,activation='relu')(fe1)

# Input 2
inputs2=Input(shape=(maximum_length,))
se1=Embedding(vocab_size,embedding_dim)(inputs2)
se2=Dropout(0.5)(se1)
se3=LSTM(256)(se2)

# decoder(feed forward) Model
decoder1=add([fe2,se3])
decoder2=Dense(256,activation='relu')(decoder1)
outputs=Dense(vocab_size,activation='softmax')(decoder2)

# Merge the two input models
model=Model(inputs=[inputs1,inputs2],outputs=outputs)
```

```python
def greedySearch(photo):
    in_text = '<startseq>'
    for i in range(maximum_length):
        sequence = [word_to_idx[w] for w in in_text.split() if w in word_to_idx]
        sequence = pad_sequences([sequence], maxlen=maximum_length)
        yhat = model.predict([photo,sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = idx_to_word[yhat]
        in_text += ' ' + word
        if word == ' <endseq>':
            break
    final = in_text.split()
    final = final[1:-1]
    final = ' '.join(final)
    return final
```
